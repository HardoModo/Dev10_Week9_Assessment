```python
import pandas as pd
from matplotlib import pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.linear_model import LassoCV
from IPython.display import display

pd.options.display.max_columns = None
pd.options.display.max_rows = 100

import numpy as np
from sklearn.impute import SimpleImputer

from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv(r"./data/titanic_train.csv")
```

```python
df.head()
y = df['Survived']
X = df.drop('Survived',axis=1)
copy= X.copy()
```

```python
display(X.head())
display(y.head())
X.info()

imp = SimpleImputer(missing_values=np.nan, strategy='mean')
numeric_cols = [cols for cols in X.columns if X[cols].dtypes != 'object']
for cols in numeric_cols:
    X[cols]=imp.fit_transform(X[[cols]])
    
```

```python
X.info()
```

```python
corr_matrix = X.corr()
print(corr_matrix)


```

```python

X[numeric_cols] = scaler.fit_transform(X[numeric_cols])

corr_matrix = X.corr()
print(corr_matrix)

display(X.head())



```

```python
X=X.drop("Cabin",axis=1)
X=X.drop("Name",axis=1)
display(X.head())
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
X['Embarked'] =imp.fit_transform(X[['Embarked']])
```

```python

X.info()

```

```python

# X=X.drop("PassengerId",axis=1)
# X=X.drop("Ticket",axis=1)
#
copy = X.copy()
cat_cols = [cols for cols in X.columns if X[cols].dtypes == 'object']

for cols in cat_cols:
    temp = pd.get_dummies(X[cols],drop_first=True)
    X = pd.concat([X,temp],axis=1)
    X = X.drop(cols,axis=1)
    print(cols)
    display(temp)
    
display(X.head())
    
```

```python
lr = LinearRegression()

Xt,Xv,yt,yv = train_test_split(X,y,test_size=0.30,random_state=0)

lr.fit(Xt,yt)

s=lr.score(Xt,yt)
t=lr.score(Xv,yv)

print(s)
print(t)

print(lr.coef_)
```

```python

las = LassoCV(cv=5,random_state=0,alphas=[0.1])
las.fit(Xt,yt)
s=las.score(Xt,yt)
t=las.score(Xv,yv)

print(s)
print(t)

table =[(x,y) for x,y in zip(las.coef_,Xt.columns)]
for t in table:
    print(f"{t[1]}: {t[0]}")
```

```python
X=copy.copy()
X=X.drop("PassengerId",axis=1)
X=X.drop("Ticket",axis=1)
#
cat_cols = [cols for cols in X.columns if X[cols].dtypes == 'object']

for cols in cat_cols:
    temp = pd.get_dummies(X[cols],drop_first=True)
    X = pd.concat([X,temp],axis=1)
    X = X.drop(cols,axis=1)
    print(cols)
    display(temp)
    
display(X.head())

Xt,Xv,yt,yv = train_test_split(X,y,test_size=0.30,random_state=0)
```

```python

newdf = pd.DataFrame()
las = LassoCV(cv=5,random_state=0,alphas=[0.1])
las.fit(Xt,yt)
s=las.score(Xt,yt)
t=las.score(Xv,yv)

print(s)
print(t)

table =[(x,y) for x,y in zip(las.coef_,Xt.columns)]
for i,cols in enumerate(Xt.columns):
    dict = {cols:[las.coef_[i]]}
    newdf = pd.concat([newdf,pd.DataFrame(dict)],axis=1)
   
newdf['Name'] = "LassoAlpha0.1"
display(newdf.head())
for t in table:
    print(f"{t[1]}: {t[0]}")

    
    


las = LassoCV(cv=5,random_state=0,alphas=[0.5])
las.fit(Xt,yt)
s=las.score(Xt,yt)
t=las.score(Xv,yv)

print(s)
print(t)
newdf2 = pd.DataFrame()
table =[(x,y) for x,y in zip(las.coef_,Xt.columns)]
for i,cols in enumerate(Xt.columns):
    dict = {cols:[las.coef_[i]]}
    newdf2 = pd.concat([newdf2,pd.DataFrame(dict)],axis=1)
   
newdf2['Name'] = "LassoAlpha0.5"

for t in table:
    print(f"{t[1]}: {t[0]}")
    
    
las = LassoCV(cv=5,random_state=0,alphas=[0])
las.fit(Xt,yt)
s=las.score(Xt,yt)
t=las.score(Xv,yv)

print(s)
print(t)

print(las.coef_)
newdf3 = pd.DataFrame()
table =[(x,y) for x,y in zip(las.coef_,Xt.columns)]
for i,cols in enumerate(Xt.columns):
    dict = {cols:[las.coef_[i]]}
    newdf3 = pd.concat([newdf3,pd.DataFrame(dict)],axis=1)
   
newdf3['Name'] = "LinearReg"

for t in table:
    print(f"{t[1]}: {t[0]}")
    
finaldf = pd.concat([newdf,newdf2],axis=0)
finaldf = pd.concat([finaldf,newdf3],axis=0)
finaldf.reset_index(inplace=True)
display(finaldf.head())

import plotly.express as px
import plotly.graph_objects as go
finaldf.index = finaldf["Name"]

finaldf= finaldf.drop(["Name","index"],axis=1)
cols = finaldf.columns
display(finaldf)


print(cols)

fig = go.Figure(data=[
    
    go.Bar(name=index, x=cols, y=row.map(lambda x: abs(x))) for index,row in finaldf.iterrows()]
                )
fig.show()
   
```
