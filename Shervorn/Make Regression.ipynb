# Make Regression Notebook

```python
from sklearn.datasets import make_regression
from matplotlib import pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
X, y = make_regression(n_samples=10000, n_features=100, noise=90, random_state=42)
```

```python
plt.hist(y);
```

## Split the data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

## Fit the Models

```python
reg = LinearRegression().fit(X_train, y_train)

ridgeModel = Ridge(alpha=1.0)
ridgeModel.fit(X_train, y_train)

lassoModel = linear_model.Lasso(alpha=1.0)
lassoModel.fit(X_train, y_train);
```

```python
lrCoef = reg.coef_
ridgeCoef = ridgeModel.coef_
lassoCoef = lassoModel.coef_

print(lassoCoef)
```

```python
lrScore = round(reg.score(X_test, y_test), 4)
ridgeScore = round(ridgeModel.score(X_test, y_test), 4)
lassoScore = round(lassoModel.score(X_test, y_test), 4)
print(f'Linear Regression Score = {lrScore}')
print(f'Ridge Score = {ridgeScore}')
print(f'Lasso Score = {lassoScore}')
```

```python
regPredict = reg.predict(X_test)
ridgePredict = ridgeModel.predict(X_test)
lassoPredict = lassoModel.predict(X_test)
```

```python
binSize = 50
```

```python
plt.hist(regPredict, bins= binSize)
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.title("Linear Regression Prediction");
```

```python
plt.hist(y_test, bins= binSize)
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.title("Actual Results");
```

```python
plt.hist(ridgePredict, bins= binSize)
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.title("Ridge Model Prediction");
```

```python
plt.hist(lassoPredict, bins= binSize)
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.title("Lasso Model Prediction");
```
